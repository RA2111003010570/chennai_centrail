```python
import folium

cities = [
    {'name':'Arakonam', 'coodrs':[13.0841, 79.6704]},
    {'name':'Arani', 'coodrs':[12.6681, 79.2847]},
    {'name':'Chennai Central', 'coodrs':[13.0832,80.2755]},
    {'name':'Chennai North', 'coodrs':[13.1626,80.2879]},
    # etc. etc.
]
m = folium.Map(zoom_start=10)
for counter, city in enumerate(cities):
    circle_size = 2 + counter
    folium.CircleMarker(
        location=city['coodrs'],
        radius=circle_size,
        popup=city['name'],
        color="crimson",
        fill=True,
        fill_color="crimson",
    ).add_to(m)

m.save('map.html')
```
